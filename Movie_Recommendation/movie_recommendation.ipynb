Importing dependencies

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and preprocessing

```python
movies_data = pd.read_csv('/content/movies.csv')
```

```python
movies_data.head()
```

```python
#number of rows of columns in the data frame

movies_data.shape
```

```python
selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
# replacing the null values with string
for features in selected_features:
  movies_data[features] = movies_data[features].fillna('')
```

```python
# combining all the 5 selected features a
combined_features = movies_data ['genres'] + '' + movies_data['keywords'] + ' ' +  movies_data['director']
```

```python
print (combined_features)
```

```python
#converting the text data to feature vector

vectorizer  = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
# getting the movie name from the user

movie_name = input('Enter your favourite movie name: ')
```

```python
from re import A
#creating a lsit with all the movie names given in the data

list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
#finding the close match for the movie name given by the user

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# Corrected code to find the index of the movie with the title close_match
index_of_the_movie = movies_data[movies_data.title == close_match].index.values[0]
print(index_of_the_movie)
```

```python
#getting the list of similar movies

similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting the movies based on their similarity score

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
# print the name of simialr movies based on index

print('Movies suggested for you: \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index == index]['title'].values[0]
  if (i<30):
    print(i ,'.',title_from_index)
    i+=1


```

Movie Recommendation System

```python
movie_name = input('Enter your favourite movie name: ')

list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title == close_match].index.values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)


print('Movies suggested for you: \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index == index]['title'].values[0]
  if (i<30):
    print(i ,'.',title_from_index)
    i+=1
```

```python
import pickle

# Assuming `similarity` is the model and `movies_data` is the DataFrame

# Save the model and data
with open('movie_recommendation_model.pkl', 'wb') as file:
    pickle.dump(similarity, file)
    pickle.dump(movies_data, file)

print("Model and data saved successfully.")
```
